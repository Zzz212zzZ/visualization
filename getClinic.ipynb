```python
import requests
import json

def get_location(name, city,pageCnt):
    ak = 'GGYY5o2G01xHP4HxgGGKuCt3yaPmOH2b'
    url = 'http://api.map.baidu.com/place/v2/search?query=' + name + f'&tag=&page_size=1000&page_num={pageCnt}&region=' + city + '&output=json&ak=' + ak
    print(url)
    response = requests.get(url)
    answer = response.json()
    print(answer['results'])
    return answer['results']


answerList=[]
for pageCnt in range(0,30):
    result=get_location('医院', '重庆市',pageCnt)
    if len(result)==0:
        break
    answerList.extend(result)




```

```python
#提取answerList，每一项为{ORGAN_NAME:name,latitude1:lat,longitude1:lng}
#最后生成outpatient
outpatient=[]
for item in answerList:
    outpatient.append({'ORGAN_NAME':item['name'],'latitude1':item['location']['lng'],'longitude1':item['location']['lat']})

#这里有一个bug，不知道为什么百度api弄下来的经纬度是反的，所以这里要把经纬度反过来
```
